```python
import pickle

with open("routes.pkl", "rb") as f:
    routes = pickle.load(f)
    print(routes)
```

```python
len(routes)
```

```python
n = len(routes)
nonDuplicateRoutes = []
for i, route in enumerate(routes):
    # print(route)
    routeSet = set()
    for station in route:
        routeSet.add(station['city'])
    # print(routeSet)
    j = i + 1
    foundDuplicate = False
    while j < n:
        otherRoute = routes[j]
        otherRouteSet = set()
        for station in otherRoute:
            otherRouteSet.add(station['city'])
        j += 1
        # check if routeSet is a subset of otherRouteSet
        if routeSet.issubset(otherRouteSet):
            print("Duplicate found: ", i, j)
            foundDuplicate = True
            break
    
    if not foundDuplicate:
        nonDuplicateRoutes.append(route)

    
```

```python
print(len(nonDuplicateRoutes))
```

```python
import networkx as nx

G = nx.DiGraph()

routes = nonDuplicateRoutes

# Populate the graph with nodes and edges from the routes
for route in routes:
    for i in range(len(route) - 1):
        start = route[i]
        end = route[i + 1]
        
        # Add nodes (cities) if they don’t already exist
        G.add_node(start['city'])
        G.add_node(end['city'])
        
        # Add directed edge with attributes
        G.add_edge(
            start['city'],
            end['city']
        )

# Output the nodes and edges
print("Nodes in the network:")
print(G.nodes(data=True))

print("\nEdges in the network:")
print(G.edges(data=True))
```

```python
nx.write_gexf(G, "train_network_duplicate_removed.gexf")
```
