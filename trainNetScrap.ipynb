```python
import requests

url = "https://railspaapi.shohoz.com/v1.0/web/train-routes"
data = {
    "model" : trainList[0]
}

response = requests.post(url, data=data)

# Print the response
print("Status Code:", response.status_code)
print("Response Content:", response.text)
```

```python
responseJson = response.json()
```

```python
import time
# post for each item in trainList and store the routes

routes = []
for train in trainList:
    data = {
        "model" : train
    }
    response = requests.post(url, data=data)
    responseJson = response.json()
    routes.append(responseJson['data']['routes'])
    # sleep
    time.sleep(.5)
```

```python
import networkx as nx

G = nx.DiGraph()

# Populate the graph with nodes and edges from the routes
for route in routes:
    for i in range(len(route) - 1):
        start = route[i]
        end = route[i + 1]
        
        # Add nodes (cities) if they don’t already exist
        G.add_node(start['city'])
        G.add_node(end['city'])
        
        # Add directed edge with attributes
        G.add_edge(
            start['city'],
            end['city']
        )

# Output the nodes and edges
print("Nodes in the network:")
print(G.nodes(data=True))

print("\nEdges in the network:")
print(G.edges(data=True))
```

```python
nx.write_gexf(G, "train_network.gexf")
```
